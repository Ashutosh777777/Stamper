```python
from openai import OpenAI
import pyttsx3
from markdown import Markdown
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"

responder = OpenAI(base_url=OLLAMA_BASE_URL, api_key="OLLAMA_API_KEY")
QUERY = input("Hey there! I am your personal assistant who can cater to your needs anytime and everytime you need me. So, what's up?")
response = responder.chat.completions.create(
    model="llama3.2",
    messages=[
        {"role":"system", "content":"You are a personal assistant that can do anything that the user would require and change your mood according to the user's"},
        {
            "role":"user", "content":f"{QUERY}"
        }                
    ]
    
)
```

```python
print(response.choices[0].message.content)
```
