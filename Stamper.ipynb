```python
print(f"Hey {name}! Type 'bye' to end the session.\n")

while True:
    QUERY = input(f"{name}: ").strip()
    if not QUERY:
        continue

    if QUERY.lower() in ("bye", "goodbye", "exit", "quit"):
        mem.log("user", QUERY)
        farewell = "Bye! It was great chatting. I'll remember this."
        mem.log("assistant", farewell)
        await Voice.speak(farewell)
        mem.end_session()
        break

    mem.log("user", QUERY)

    long_term  = mem.retrieve_context(QUERY)
    short_term = mem.get_recent_turns(n=10)

    response = responder.chat.completions.create(
        model="llama3.2",
        messages=[
            {"role": "system", "content": (
                f"You are a personal assistant that can do anything that the user would require "
                f"and change your mood according to the user's. The user has {hobbies} and you must "
                f"talk to the user according to her/his hobbies so that they feel comfortable talking "
                f"to you, but not have references in each line, that would make the user uncomfortable. "
                f"You have to be a good friend to the user, that may involve talking references from "
                f"movies, books, telling the user truth about their choices, occasional teasing, etc. "
                f"Keep your answers simple and short where the question is straightforward, keep it "
                f"spicy when it is complex for the user to better understand it. But remember, your "
                f"provided output will be passed to the user in voice format, so make sure to provide "
                f"the output in a speakable format. Not in a code format, not in a markdown format, "
                f"just text and no specification of your mood is required in the output.\n\n"
                + long_term
            )},
            *short_term,
            {"role": "user", "content": QUERY}
        ]
    )

    reply = response.choices[0].message.content
    mem.log("assistant", reply)
    print(f"Natasha: {reply}")
    await Voice.speak(reply)
```
